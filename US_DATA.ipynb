```python
# Drop duplicates based on the 'year' column, keeping the first occurrence
df_unique_years = df.drop_duplicates(subset=['YEAR'], keep='first')

df_unique_years
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['ANN'])
plt.title('Temperature in New York')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

```python
# Create the histogram
plt.hist(df['ANN'], bins=20, color='cyan', edgecolor='black')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.grid(True)
plt.show()
```

```python
# Load the shapefile using geopandas
gdf = gpd.read_file("gadm41_USA_1.shp")
fig, ax = plt.subplots(figsize=(40, 40))
gdf.plot(ax=ax, color='lightgray', edgecolor='black')
plt.show()
```

```python
# Load the shapefile using geopandas
gdf = gpd.read_file("gadm41_USA_1.shp")
fig, ax = plt.subplots(figsize=(40, 40))
gdf.plot(ax=ax, color='lightgray', edgecolor='black')
# Remove axes
ax.set_axis_off()
plt.title('USA Map (Enlarged)', fontsize=20)#adjusted fontsize
plt.show()
```
