```python
import pandas as pd
import matplotlib.pyplot as plt 
import geopandas as gpd
```

```python
df = pd.read_csv("./canada_data.csv")#contains 3 parameters
df
```

```python
df.head(19)#only temperature
```

```python
df.describe()
```

```python
# Drop duplicates based on the 'year' column, keeping the first occurrence
df_unique_years = df.drop_duplicates(subset=['YEAR'], keep='first')

df_unique_years
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['YEAR'], df['ANN'])
plt.title('Temperature in canada')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
```

```python
# Create the histogram
plt.hist(df['ANN'], bins=15, color='magenta', edgecolor='black')
plt.xlabel(' annual average temperature')
plt.ylabel('count')
plt.title('Temperature Distribution')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(df['ANN'], color='magenta', fill=True)  # Shade fills the area under the curve
plt.xlabel('Temperature')
plt.ylabel('Density')
plt.title('Temperature Distribution')
plt.xlim(-40, 20)  # Set x-axis limits for desired range
plt.grid(True)
plt.show()
```
